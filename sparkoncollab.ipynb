```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").appName("ColabSpark").getOrCreate()
spark
```

```python
!wget -O yellow_tripdata_2023-01.parquet https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet
```

```python
df = spark.read.parquet("yellow_tripdata_2023-01.parquet")
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
columns_to_drop = ["airport_fee", "ehail_fee"]
df = df.drop(*columns_to_drop)
```

```python
df = df.na.drop()
```

```python
from pyspark.sql.functions import col

df = df.withColumn("tpep_pickup_datetime", col("tpep_pickup_datetime").cast("timestamp"))
df = df.withColumn("tpep_dropoff_datetime", col("tpep_dropoff_datetime").cast("timestamp"))
```

```python
df.count()
```

```python
from pyspark.sql.functions import count

df.groupBy("payment_type").agg(count("*").alias("count")).orderBy("count", ascending=False).show()
```

```python
from pyspark.sql.functions import hour, avg

df.groupBy(hour("tpep_pickup_datetime").alias("hour")) \
  .agg(avg("fare_amount").alias("average_fare")) \
  .orderBy("hour").show()
```

```python
df.groupBy("PULocationID").agg(count("*").alias("num_trips")) \
  .orderBy("num_trips", ascending=False).show(10)
```

```python
df.cache()
```

```python
df.write.mode("overwrite").parquet("nyc_taxi_cleaned.parquet")
```

```python
from pyspark.sql.functions import col

df.orderBy(col("trip_distance").desc()).show(1)
```

```python
df.groupBy(hour("tpep_pickup_datetime").alias("hour")) \
  .agg(count("*").alias("num_trips")) \
  .orderBy("hour").show()
```

```python
df.write.csv("nyc_taxi_results.csv", header=True, mode="overwrite")
```
